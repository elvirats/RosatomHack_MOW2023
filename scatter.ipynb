```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('scatter.csv')
```

```python
fig = plt.subplots(figsize=(15, 10))
outliers = data.loc[data.labels == -1, :]
clustered = data.loc[data.labels != -1, :]

plt.scatter(outliers.feat_0, outliers.score, color='#BDBDBD', s=50)
plt.scatter(clustered.feat_0, clustered.score, c=clustered.labels, s=50, cmap='hsv_r')
# fig.set_xlabel("", fontsize=15, color='blue', labelpad=120)    # +
# fig.set_ylabel("", fontsize=15, color='orange', labelpad=140) 
```
